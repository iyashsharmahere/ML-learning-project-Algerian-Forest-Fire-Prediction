```python
df=pd.read_csv('Algerian_forest_fires_dataset_cleaned.csv')
```

```python
df.head()
```

```python
df=df.drop(['month','day','year'],axis=1)
```

```python
df.head()
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df['Classes'].value_counts()
```

```python
df.columns
```

```python
X=df.drop(['FWI'],axis=1)
y=df['FWI']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
def correlation(dataset,threshold):
  col_corr=set()
  col_matrix=dataset.corr()
  for i in range(len(col_matrix.columns)):
    for j in range(i):
      if abs(col_matrix.iloc[i,j])>threshold:
        colname=col_matrix.columns[i]
        col_corr.add(colname)
  return col_corr
```

```python
correlation(X_train,0.85)
```

```python
X_train.drop(['BUI', 'DC'],axis=1,inplace=True)
X_test.drop(['BUI', 'DC'],axis=1,inplace=True)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_scaled=sc.fit_transform(X_train)
X_test_scaled=sc.transform(X_test)
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('Before Scaling')
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title('After Scaling')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
lr=LinearRegression()
lr.fit(X_train_scaled,y_train)
```

```python
y_pred=lr.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)
```

```python
print(mae,mse,rmse,r2)
```

```python
plt.scatter(y_test,y_pred)
```

```python
#lasso regression
from sklearn.linear_model import Lasso
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred_lasso=lasso.predict(X_test_scaled)
mae_lasso=mean_absolute_error(y_test,y_pred_lasso)
mse_lasso=mean_squared_error(y_test,y_pred_lasso)
rmse_lasso=np.sqrt(mse_lasso)
r2_lasso=r2_score(y_test,y_pred_lasso)
print(mae_lasso,mse_lasso,rmse_lasso,r2_lasso)
plt.scatter(y_test,y_pred_lasso)
```

```python

from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred_ridge=ridge.predict(X_test_scaled)
mae_ridge=mean_absolute_error(y_test,y_pred_ridge)
mse_ridge=mean_squared_error(y_test,y_pred_ridge)
rmse_ridge=np.sqrt(mse_ridge)
r2_ridge=r2_score(y_test,y_pred_ridge)
print(mae_ridge,mse_ridge,rmse_ridge,r2_ridge)
plt.scatter(y_test,y_pred_ridge)
```

```python
#elastic net regression
from sklearn.linear_model import ElasticNet
elastic=ElasticNet(l1_ratio=0.2)
elastic.fit(X_train_scaled,y_train)
y_pred_elastic=elastic.predict(X_test_scaled)
mae_elastic=mean_absolute_error(y_test,y_pred_elastic)
mse_elastic=mean_squared_error(y_test,y_pred_elastic)
rmse_elastic=np.sqrt(mse_elastic)
r2_elastic=r2_score(y_test,y_pred_elastic)
print(mae_elastic,mse_elastic,rmse_elastic,r2_elastic)
plt.scatter(y_test,y_pred_elastic)
```

```python
from sklearn.linear_model import LassoCV,RidgeCV,ElasticNetCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
y_pred_lassocv=lassocv.predict(X_test_scaled)
mae_lassocv=mean_absolute_error(y_test,y_pred_lassocv)
r2_lassocv=r2_score(y_test,y_pred_lassocv)
print(mae_lassocv,r2_lassocv)
plt.scatter(y_test,y_pred_lassocv)
```
